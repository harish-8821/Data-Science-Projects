```python
import pandas as pd
```

```python
df=pd.read_csv("heart.csv")
df
```

```python
df.HeartDisease.value_counts()
```

```python
dummies = pd.get_dummies(df['Sex'])
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
dummies = pd.get_dummies(df['ChestPainType'])
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
final=merged.drop(['ChestPainType','RestingECG','ExerciseAngina','MaxHR','FastingBS'],axis='columns')
final
```

```python
du=pd.get_dummies(df.Sex)
du
```

```python
me=pd.concat([final,du],axis='columns')
me
```

```python
d=pd.get_dummies(me.ST_Slope)
d
```

```python
m=pd.concat([me,d],axis='columns')
m
```

```python
End=m.drop(['Sex','ST_Slope'],axis='columns')
End
```

```python
End.M=End.M.fillna("0")
```

```python
X = End.drop("HeartDisease",axis="columns")
y = End.HeartDisease
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
y_train.value_counts()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
